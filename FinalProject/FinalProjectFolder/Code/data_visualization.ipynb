```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('../Data/reddit_posts_with_sentiment.csv')
df['created_utc'] = pd.to_datetime(df['created_utc'])

# Sentiment Over Time
df.set_index('created_utc', inplace=True)
sentiment_over_time = df['sentiment'].resample('W').mean()

plt.figure(figsize=(10, 5))
plt.plot(sentiment_over_time)
plt.title('Sentiment Over Time')
plt.xlabel('Time')
plt.ylabel('Average Sentiment Score')
plt.savefig('../Images/sentiment_over_time.png')

# Sentiment Intensity
plt.figure(figsize=(10, 5))
plt.hist(df['sentiment'], bins=200, edgecolor='black')
plt.title('Sentiment Intensity')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.savefig('../Images/sentiment_intensity.png')

# Volume of Posts
post_volume = df.resample('W').size()

plt.figure(figsize=(10, 5))
plt.bar(post_volume.index, post_volume.values)
plt.title('Volume of Posts')
plt.xlabel('Time')
plt.ylabel('Number of Posts')
plt.savefig('../Images/volume_of_posts.png')
```
