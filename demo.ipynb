```python
import nltk
import pandas as pd
nltk.download("all")
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Define the text to be preprocessed
text = 'Example sentence needing preprocessing. There are lots of dogs running around. The dogs are running around the neighborhood. I saw a dog at the park.'

# Tokenize the text
tokens = word_tokenize(text.lower())

# Remove stopwords
stopwords_list = stopwords.words('english')
cleaned_tokens = [word for word in tokens if word not in stopwords_list]

# Lemmatize the tokens
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in cleaned_tokens]

# Print the cleaned and lemmatized tokens
print(lemmatized_tokens)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
sample_text = ["This is the first document.", "This document is the second document."]
vectors = vectorizer.fit_transform(sample_text)
count_vector_df = pd.DataFrame(vectors.todense(), columns=vectorizer.get_feature_names_out())
count_vector_df.head()
```
